```sql
-- Daily Stock Screener

with latest_data as (
  select *
  from analytics.mart_stock_features
  where trade_date = (
    select max(trade_date)
    from analytics.mart_stock_features
  )
),

filtered as (
  select *
  from latest_data
  where 
    -- Core breakout condition
    is_breakout_candidate = 1
    -- Trend confirmation
    and return_20d > 0.05
    -- Volume confirmation
    and is_volume_spike = 1
    -- Avoid extremely volatile names
    and vol_20d < 0.5 
)

select 
  ticker,
  trade_date,
  close,
  return_5d,
  return_20d,
  return_50d,
  vol_20d,
  avg_volume_20d,
  pct_of_20d_high,
  momentum_rank
from filtered
order by return_20d desc
limit 20;
```
